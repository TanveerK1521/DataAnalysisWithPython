### DATA NORMALIZATION  
Is the process of transforming values of several variables into similar range.  
A typical normalization include:  
Scaling the variable so that variable average is 0  
Scaling the variable so that variable variance is 1  
Scaling the variable so that varibale range is between 0 and 1  
To demonstarte normalization say we want to scale 'length', 'width' and height  
Target is to normalize there values between 0 and 1  
Approach is to replace original value of original/ max

```python
# replace (original value) by (original value)/(maximum value)
df2['length'] = df['length']/df['length'].max()
df2['width'] = df['width']/df['width'].max()
df2['height'] = df['height']/df['height'].max()
```

### BINNING  
Is the process of transforming continous numerical variables into discrete categorical bins for grouped analysis.  
For example if you want to find price difference between cars with high horsepower, medium and low. We can rearrange them in three bins to simplfy analysis.

```python
df2['horsepower'] = df2['horsepower'].astype(int, copy = True)
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df2["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
# Now find three bins of equal band width using Numpy's linspace method
bins = np.linspace(min(df2['horsepower']),max(df2['horsepower']),4)

#setting group names
group_names = ['Low', 'Medium', 'High']

# then apply the function cut to figure out what each value of column belongs to which bin
df2['horsepower-binned'] = pd.cut(df2['horsepower'], bins, labels=group_names, include_lowest=True )
df2[['horsepower','horsepower-binned']].head(20)
```

CORRELATION AND CAUSATION  
Correlation is measure of extent of interdependance between variables.  
Causation is relationship between cause and effect between two variables.  
#### Pearson Correlation  
Pearson Correlation measures linear dependance between X and Y  
The resulting coefficient is a value between -1 and 1.  
Where 1 means perfect positive linear correlation, 0 means no correlation and -1 means perfect negative correlation  
Pearson correlation is a default method of function corr().  
SOmetimes we need to know the significance of correlation estimate  
P- Value: Is the probabilty value that correlation between these two variables is statistically significant.  
Normally we choose 0.05 significance level which means we are 95 percent confident that correlation between variable is significant  
If p-value < 0.001 then we have a strong evidence that the correlation is significant  
If p-value<0.05 then moderate evidance  
If p-value<0.1 Weak evidance  
If p-value> 0.1 no evidence that correlation is significant or not  
We can obtain this information using stats module in spicy library

```python
from scipy import stats
pearson_coef, p_value = stats.pearsonr(df['wheel-base'], df['price'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)  
```
